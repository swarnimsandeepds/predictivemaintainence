```python
import matplotlib.pyplot as plt

# Assuming 'TWF', 'HDF', 'PWF', 'OSF', 'RNF' columns exist in your DataFrame
failure_modes = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']

# Number of machine types
num_machine_types = len(df['Type'].unique())

# Set up subplots in a single row
fig, axes = plt.subplots(1, num_machine_types, figsize=(16, 6))  # Adjust figsize as needed

# Iterate over unique machine types ('L', 'M', 'H')
for i, machine_type in enumerate(df['Type'].unique()):
    # Filter DataFrame for the current machine type
    df_machine_type = df[df['Type'] == machine_type]

    # Count failures for the current machine type
    failure_counts_machine_type = df_machine_type[failure_modes].sum()

    # Plot a pie chart for failures in the current machine type
    axes[i].pie(failure_counts_machine_type, labels=failure_counts_machine_type.index, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Distribution of Failures for Type {machine_type}')

plt.show()
```

```python

# Assuming 'TWF', 'HDF', 'PWF', 'OSF', 'RNF' columns exist in your DataFrame
failure_modes = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']

# Number of machine types
num_machine_types = len(df['Type'].unique())

# Set up subplots in a single row
fig, axes = plt.subplots(1, num_machine_types, figsize=(16, 6))  # Adjust figsize as needed

# Iterate over unique machine types ('L', 'M', 'H')
for i, machine_type in enumerate(df['Type'].unique()):
    # Filter DataFrame for the current machine type
    df_machine_type = df[df['Type'] == machine_type]

    # Count failures for the current machine type
    failure_counts_machine_type = df_machine_type[failure_modes].sum()

    # Plot a pie chart for failures in the current machine type
    wedges, texts, autotexts = axes[i].pie(failure_counts_machine_type, labels=failure_counts_machine_type.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(edgecolor='black'))

    # Make all text bold
    for text in texts + autotexts:
        text.set_weight('bold')

    # Set the title fontweight to bold
    axes[i].set_title(f'Distribution of Failures for Type {machine_type}', fontweight='bold')

    axes[i].legend(failure_counts_machine_type.index, loc='upper right')
    axes[i].set_aspect('equal')

# Adjust layout
plt.tight_layout()

plt.show()
```
